# Question - 3

```python
import statistics
from math import sqrt
x = np.linspace(-5, 5, 100)
mean = 0
variance = 1
# the formula for uniform distribution range [a, b]
a = mean - variance*(sqrt(3))
b = mean + variance*(sqrt(3))
init_point = a
distance = b-a
normal_values =norm.pdf(x, mean, variance)
uniform_values = uniform.pdf(x, init_point, distance) # b-a is the distance from the inital point till which p(x) â‰  0
```

```python
plot.title("Question - 3")
plot.plot(x, normal_values, label='Normal Distribution')
plot.plot(x, uniform_values, label = 'Uniform Distribution')
plot.legend()
plot.show()
```

We know that the peak of the normal distribution represents the mean of the distribution. We can see that the peak is close to 250 from the graph. One can conclude that the distribution of the given random variable is normal. This is further proven with the mean of the distribution which is approx. 250.    

```python
from math import exp, log
from statistics import NormalDist
class main_object():
    def __init__(self):
        self.data = [random.random() for _ in range(10000)]
    def rayleigh_inverse(self, sigma):
        return [sigma*(sqrt(2))*(sqrt(log(1/(1-i)))) for i in self.data]
    def exp_inverse(self, lam):
        return [(1/lam)*log(1/ (1 - i)) for i in self.data]
    def normal_inverse(self, mean, sigma):
        return [NormalDist(mu = mean, sigma=sigma).inv_cdf(i) for i in self.data]
q6_object = main_object()
```

```python
q6_rayleigh = q6_object.rayleigh_inverse(sigma=1.0)
q6_exp = q6_object.exp_inverse(lam=1.5)
q6_normal = q6_object.normal_inverse(mean=0, sigma=3.0)
```

## Normal Plot

```python
plot.title("Normal Plot")
plot.hist(q6_normal, bins=100, density=True)
index_array = np.linspace(-15, 15, 1000)
plot.plot(index_array, norm.pdf(index_array, 0, 3))
plot.show()
```

## Rayleigh Plot

```python
from scipy.stats import rayleigh
plot.title("Rayleigh Plot")
plot.hist(q6_rayleigh, bins=100, density=True)
index_array = np.linspace(0, 10, 1000)
plot.plot(index_array, rayleigh.pdf(index_array, 0, 1))
plot.show()
```

## Exponential Plot

```python
from scipy.stats import expon
plot.title("Exponential Plot")
plot.hist(q6_exp, bins = 100, density=True)
index_array = np.linspace(0, 7, 1000)
plot.plot(index_array, expon.pdf(index_array, 0, 1))
plot.show()
```
